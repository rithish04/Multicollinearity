```python
import pandas as pd
```

```python
import statsmodels.api as sm
df_adv = pd.read_csv('Advertising.csv',index_col=0)
x = df_adv[['TV','Radio','Newspaper']]
y = df_adv['Sales']
df_adv.head()
```

```python
#fit a ols model with with intercept on tv and radio
x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```
